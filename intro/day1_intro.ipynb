### Bias
Sometimes data might be skewed or might be in favor of certain individuals. For example, in a survey about weather in the US, if you only survey people from California, you will have biased results compared to surveying people from different states.

# What is Machine Learning and Artificial intelligence? 

**Machine Learning** is **Artificial Intelligence**. **Artificial Intelligence** is **Machine Learning**

These are "fields" or sectors within the Computer Science and Data Science world. ML and AI are built of algorithms and designs that are made to automate the work being done with data and computers. 

These include a varitey of algorithms like supervised vs unsupervised model, from which the computer is learning something given information and data. AI makes use of ML to take in data and information and take actions based on that. 

<br>
<img src='images/ml.jpg' width="500" height="500"/></a>
<center>Source: infoworld.com</center>
<br>


### Sub-topic is Classification/Prediction: 

For example, you recieve emails that are **spam** (bad) and **ham** (good). 

The **problem** is that you don't know which emails are good and bad. 

In order to **predict** or **classify** emails into "good" and "bad" you can build a Machine Learning model that can make these groups for you.
